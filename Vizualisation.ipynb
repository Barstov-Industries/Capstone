```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
barstov_merged = pd.read_csv('../Capstone/Data/barstov_merged.csv', sep=',')
```

```python
#Only 2018

# Convert 't_dat' to datetime, specifying the format
barstov_merged['t_dat'] = pd.to_datetime(barstov_merged['t_dat'], format='%Y-%m-%d')

# Filter for the year 2018
data_2018 = barstov_merged[barstov_merged['t_dat'].dt.year == 2018]

# Group by date and sum the income (price) for each day
income_2018 = data_2018.groupby('t_dat')['price'].sum().reset_index()

# Plotting the line plot
plt.figure(figsize=(14, 6))
sns.lineplot(data=income_2018, x='t_dat', y='price')
plt.title("Sales Over Time in 2018")
plt.xlabel("Date")
plt.ylabel("Total Income")
plt.xticks(rotation=45)
plt.show()
```
